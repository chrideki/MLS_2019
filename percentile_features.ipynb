```python
import numpy as np
import pandas as pd
import re
import datetime
from datetime import date
import dateutil
```

```python
data = pd.read_csv('mls_player.csv')
```

```python
# to show all columns' names 
# str(data.columns.tolist())
```

```python
data['MP_Pct'] = data['Playing Time_MP'].rank(pct=True)

# defensive features for radar chart
data['Performance_CrdY90'] = data['Performance_CrdY'] / data['Playing Time_90s']
data['CrdY_Pct'] = data['Performance_CrdY90'].rank(pct=True)
data['Aerial Duels_Won90'] = data['Aerial Duels_Won'] / data['Playing Time_90s']
data['Aerial Duels_Won_Pct'] = data['Aerial Duels_Won90'].rank(pct=True)
data['Aerial Duels_Won%_Pct'] = data['Aerial Duels_Won%'].rank(pct=True)
data['Tackles_TklW90'] = data['Tackles_TklW'] / data['Playing Time_90s']
data['TklW90_Pct'] = data['Tackles_TklW90'].rank(pct=True)
data['Dribbles_Tkl%_Pct'] = data['Vs Dribbles_Tkl%'].rank(pct=True)
data['Blocks_Blocks90'] = data['Blocks_Blocks'] / data['Playing Time_90s']
data['Blocks90_Pct'] = data['Blocks_Blocks90'].rank(pct=True)
```

```python
# passing and possession features for radar chart
data['Passing%_Pct'] = data['Total_Cmp%'].rank(pct=True)
data['Pass Types_TB90'] = data['Pass Types_TB'] / data['Playing Time_90s']
data['TB90_Pct'] = data['Pass Types_TB90'].rank(pct=True)
data['Pass Types_Sw90'] = data['Pass Types_Sw'] / data['Playing Time_90s']
data['Sw90_Pct'] = data['Pass Types_Sw90'].rank(pct=True)
data['Pass Types_Crs90'] = data['Pass Types_Crs'] / data['Playing Time_90s']
data['Crs90_Pct'] = data['Pass Types_Crs90'].rank(pct=True)
data['Dribbles_Succ%_Pct'] = data['Dribbles_Succ%'].rank(pct=True)
data['Dispos90'] = data['Dispos'] / data['Playing Time_90s']
data['Dispos90_Pct'] = data['Dispos90'].rank(pct=True)
```

```python
# attacking features for radar chart
data['SCA90_Pct'] = data['SCA_SCA90'].rank(pct=True)
data['GCA90_Pct'] = data['GCA_GCA90'].rank(pct=True)
data['Ast90_Pct'] = data['Per 90 Minutes_Ast'].rank(pct=True)
data['Gls90_Pct'] = data['Per 90 Minutes_Gls'].rank(pct=True)
```

```python
col_list = ['ID','name', 'Age', 'position', 'Squad']
drop_list = ['CrdY_Pct', 'Aerial Duels_Won_Pct', 'Dribbles_Tkl%_Pct', 'Sw90_Pct', 'Dispos90_Pct' ]
for col in data.columns:
    if 'Pct' in col and col not in drop_list:
        col_list.append(col)
len(col_list)   
```

```python
# we consider only performances in MLS
mls_data = data[data['Comp'] == '1. MLS'] 
```

```python
mls_data.to_csv ('mls_2019.csv', index = False, header=True)
```
